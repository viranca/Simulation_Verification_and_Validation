```python
dfnodes = pd.read_csv('A320nodes.csv')
dfdefl = pd.read_csv('ULC1.csv')
```

```python
slc1 = pd.read_csv('SLC1.csv')
slc2 = pd.read_csv('SLC2.csv')
slc3 = pd.read_csv('SLC3.csv')
slc4 = pd.read_csv('SLC4.csv')
slc1 = slc1.drop(columns=['Element Label'], axis=1)
slc2 = slc2.drop(columns=['Element Label'], axis=1)
slc3 = slc3.drop(columns=['Element Label'], axis=1)
slc4 = slc4.drop(columns=['Element Label'], axis=1)
"""
slc1 = slc1.set_index('Node Label')
slc2 = slc2.set_index('Node Label')
slc3 = slc3.set_index('Node Label')
slc4 = slc4.set_index('Node Label')
"""
slc1
```

```python
node_label = np.arange(1, 6555)
misses = DataFrame(node_label)
#misses['misses1'] = 0
#misses['misses2'] = 0
#misses['misses3'] = 0
#misses['misses4'] = 0
#misses['misses5'] = 0
#misses['misses6'] = 0
#misses['misses7'] = 0
#misses['misses8'] = 0
misses.columns = ['Node Label']
misses    
```

```python
misses = pd.merge(misses, slc1, how='left', on=['Node Label'])
misses = pd.merge(misses, slc2, how='left', on=['Node Label'])
misses = pd.merge(misses, slc3, how='left', on=['Node Label'])
misses = pd.merge(misses, slc4, how='left', on=['Node Label'])
misses
```

```python
for i in range(len(misses.index)-1, 1, -1):
    if misses.values[i,0] == misses.values[i-1 ,0]:
        misses = misses.drop(misses.index[i])

misses2=misses
```

```python
misses=misses2
misses.columns
```

```python
misses['summ'] = 0

misses['summ'] = misses.Mises1.fillna(0) + misses.fillna(0)['Mises2'] + misses.fillna(0)['Mises3'] + misses.fillna(0)['Mises4'] + misses.fillna(0)['Mises5'] + misses.fillna(0)['Mises6'] + misses.fillna(0)['Mises7'] + misses.fillna(0)['Mises8']

```

```python
misses['numberamount'] = (misses.apply(lambda x: x.count(), axis=1)-3)
```

```python
misses['missesavg'] = misses['summ']/misses['numberamount']
```

```python
misses
```

```python
misses = misses.drop(columns=['Mises1', 'Mises2','Mises3', 'Mises4','Mises5', 'Mises6','Mises7', 'Mises8', 'numberamount', 'summ'])
```
